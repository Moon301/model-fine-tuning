```python
import pandas as pd
import os

# 합칠 파일 경로
csv1_path = "/mnt/hdd1/moon/ev_data/dataset/betterwhy_data_label.csv"
csv2_path = "/mnt/hdd1/moon/ev_data/dataset/sk_data_label.csv"

# 출력 파일 경로
output_path = "/mnt/hdd1/moon/ev_data/dataset/merged_dataset.csv"

# CSV 파일 불러오기
df1 = pd.read_csv(csv1_path)
df2 = pd.read_csv(csv2_path)

# 두 데이터프레임 합치기
merged_df = pd.concat([df1, df2], ignore_index=True)

# 합친 데이터 저장
merged_df.to_csv(output_path, index=False)

print(f"✅ CSV 파일 합치기 완료! 저장 경로: {output_path}")
```
